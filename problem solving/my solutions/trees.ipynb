## **MAX PATH SUM**

```
def algorithm(node):
  # base case
  if node == None:
    return 0, 0
  # recursive step
  left_max_sum, left_sum = algorithm(node.left)
  right_max_sum, right_sum = algorithm(node.right)
  tmp = node.value + left_sum + right_sum
  max_sum = max(left_max_sum, right_max_sum)
  if tmp > max_sum:
    max_sum = tmp
  return max_sum, node.value + max(left_sum, right_sum)
```

```
def postorder(root):
  stack = [root]
  prev = None
  while len(stack) > 0:
    curr = stack[-1] # peek
    # leaf node
    if curr.left == None and curr.right == None:
      prev = stack.pop()
      visit_node(prev)
      continue
    # none-leaf node
    if prev == curr.right:
      # this node's children were visited
      prev = stack.pop()
      visit_node(prev)
      continue
    if not (curr.right == None):
      stack.append(curr.right)
    if not (curr.left == None):
      stack.append(curr.left)
```

## **Check Balance**

```
import Math
def algorithm(node):
  # base case
  if node == None:
    return 0
  # recursive step
  left_height = algorithm(node.left)
  right_height = algorithm(node.right)
  curr_height = 1 + max(left_height, right_height)
  if Math.abs(left_height - right_height) > 1:
    return curr_height, False
  else:
    return curr_height, True
```

## **Symmetric Tree**

```
def algorithm(node1, node2):
  # base case
  if node1 == None and node2 == None:
    return True
  # recursive step
  return (node1.value == node2.value) and algorithm(node1.left, node2.right) and algorithm(node1.right, node2.left)
```

## **BST Iterator**

```
class BSTIterator:
  # attributes
  stack = []
  index = 0
  # contructor
  def __init__(self, root):
    curr = root
    while curr != None:
      self.stack.push(curr)
      curr = curr.left
  # methods
  def next(self):
    tmp = self.stack.pop()
    index += 1
    curr = tmp.right
    while curr != None:
      self.stack.push(curr)
      curr = curr.left
    return tmp
  
  def hasNext(self):
    return len(self.stack) > 0
```

### **Lowest Common Ancestor**

```
def algorithm(node, val1, val2, result):
  # base case
  if node == None:
    return False
  # recursive step
  left = algorithm(node.left,val1, val2, result)
  right = algorithm(node.right,val1, val2, result)
  curr = node.value == val1 or node.value == val2
  if (left and right) or (curr and left) or (curr and right):
    result["result"] = node
  return left or right or curr
```

## **Largest BST subtree**

```
def algorithm(node, left_limit, right_limit):
  # base case
  if node == None:
    return 1, 0
  if node.left == None and node.right == None:
    return 1, 1
  # recursive step
  l_bst, l_count = algorithm(node.left, left_limit, node.value)
  r_bst, r_count = algorithm(node.right, node.value, right_limit)
  if l_bst and r_bst and node.value > left_limit and node.value <= right_limit
    return 1, l_count + r_count + 1
  return 0, max(l_count, r_count)
```

```
def algorithm(arr, start, end, mem): # start and end inclusive
  # base case
  if start > end:
    return 1
  # recursive step
  n = end - start + 1
  if n in mem:
    return mem[n]
  sum = 0
  for i in range(start, end + 1):
    l_count = algorithm(arr, start, i - 1, mem)
    r_count = algorithm(arr, i + 1, end, mem)
    # calculate count
    if l_count == 0:
      sum = sum + r_count
    elif r_count == 0:
      sum = sum + l_count
    else:
      sum = sum +  (l_count * r_count)
  mem[n] = sum
  return sum
```
