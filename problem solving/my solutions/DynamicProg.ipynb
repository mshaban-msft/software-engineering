Palindrome Partitioning

```
def isPalindrome(s, start, end):
  i = start
  j = end
  while i < j:
    if not s[i] == s[j]:
      return False
    i += 1
    j -= 1
  return True
```

```
def algorithm(s, start, curr):
  # base case
  if curr >= len(s):
    return False, [[]]
  if start == curr and start == len(s) - 1:
    return True, [[s[start]]]
  # recursive step
  if isPalindrome(s, start, curr):
    cond1, lists1 = algorithm(s, curr + 1, curr + 1)
    for l in lists1:
      l.insert(0, s[start: curr+1])
    cond2, lists2 = algorithm(s, start, curr + 1)
    if cond2:
      return True, lists1 + lists2
    else:
      return True, lists1
  else:
    return algorithm(s, start, curr + 1)
```

```
def algorithm(s, start, curr, mem):
  # base case
  if curr >= len(s):
    return False, [[]]
  if mem[start] is not None:
    return True, mem[start]
  if start == curr and start == len(s) - 1:
    mem[start] = [[s[start]]]
    return True, mem[start]
  # recursive step
  if isPalindrome(s, start, curr):
    cond1, lists1 = algorithm(s, curr + 1, curr + 1, mem)
    lists11 = [row[:] for row in lists1]
    for l in lists11:
      l.insert(0, s[start: curr+1])
    cond2, lists2 = algorithm(s, start, curr + 1, mem)
    res = []
    if cond2:
      res = lists11 + lists2
    else:
      res = lists11
    if start == curr:
      mem[start] = res
    return True, res
  else:
    return algorithm(s, start, curr + 1, mem)
```

```
def algorithm(s, start, curr, mem):
  # base case
  if curr >= len(s):
    return False, 0
  if mem[start] is not None:
    return True, mem[start]
  if start == curr and start == len(s) - 1:
    mem[start] = 1
    return True, mem[start]
  # recursive step
  if isPalindrome(s, start, curr):
    cond1, count1 = algorithm(s, curr + 1, curr + 1, mem)
    cond2, count2 = algorithm(s, start, curr + 1, mem)
    res = count1
    if cond2:
      res += count2
    if start == curr:
      mem[start] = res
    return True, res
  else:
    return algorithm(s, start, curr + 1, mem)
```

```
def algorithm(arr, sum, i):
  # base case
  if sum == 0:
    return 0
  if i >= len(arr):
    return float("inf")
  # recursive step
  return min(
      1 + algorithm(arr, sum - arr[i], i + 1) if sum - arr[i] >= 0 else float("inf"),
      algorithm(arr, sum, i + 1)
  )
```

```
arr = [1,2,5]
sum = 7
res = algorithm(arr, sum, 0)
print(res)
```

## **Buy and Sell Stock**

```
def algorithm(prices, start, k, mem):
  # base case
  if k == 0 or start == len(prices):
    return 0
  # memoization
  if mem[start][k - 1] is not None:
    return mem[start][k - 1]
  # recursive step
  low = prices[start]
  best = 0
  curr = 0
  for i in range(start + 1, len(prices)):
    if prices[i] <= low:
      low = prices[i]
    elif prices[i] > low:
      curr = prices[i] - low
    tmp = curr + algorithm(prices, i + 1, k - 1, mem)
    if tmp > best:
      best = tmp
  mem[start][k - 1] = best
  return best
```

```
prices = [1, 4, 5, 7, 6, 3, 2, 9, 321, 4646, 111, 99, 3, 11, 44, 55, 66, 1, 1, 2, 66, 77, 55]
n = len(prices)
k = 2
mem = [[None for i in range(k)] for j in range(n)] # n * k matrix
res = algorithm(prices, 0, k, mem)
print(res)
```

## **Longest common subsequence**

```
def algorithm(str1, str2, i, j):
  # base case
  if i == len(str1) or j == len(str2):
    return 0
  # recursive step
  if str1[i] == str2[j]:
    return 1 + algorithm(str1, str2, i + 1, j + 1)
  else:
    return max(
        algorithm(str1, str2, i + 1, j),
        algorithm(str1, str2, i, j + 1)
    )
```

```
str1 = "abcde"
str2 = "ace"
res = algorithm(str1, str2, 0, 0)
print(res)
```

```
def algorithm(str1, str2):
  # init
  m = len(str1)
  n = len(str2)
  mem = [[0 for i in range(n)] for j in range(m)] # m * n matrix
  # bottom-up
  for i in range(0, m):
    for j in range(0, n):
      if str1[i] == str2[j]:
        if i == 0 or j == 0:
          mem[i][j] = 1
        else:
          mem[i][j] = 1 + mem[i-1][j-1]       
      else:
        if i == 0:
          mem[i][j] = mem[i][j-1]
        elif j == 0:
          mem[i][j] = mem[i-1][j]  
        else:
          mem[i][j] = max(mem[i-1][j], mem[i][j-1])
  # return 
  return mem[m-1][n-1]
```

```
str1 = "abcde"
str2 = "ace"
res = algorithm(str1, str2)
print(res)
```

## **Longest Common Substring**

```
def algorithm(str1, str2, i, j):
  # base case
  if i == len(str1) or j == len(str2):
    return 0
  # recursive step
  if str1[i] == str2[j]:
    currLen = findCommonLength(str1, str2, i, j)
    return max(
        currLen,
        algorithm(str1, str2, i + 1, j),
        algorithm(str1, str2, i, j + 1)
    )
  else:
    return max(
        algorithm(str1, str2, i + 1, j),
        algorithm(str1, str2, i, j + 1)
    )
```

```
def findCommonLength(str1, str2, i, j):
  res = 0
  while str1[i] == str2[j]:
    res += 1
    i += 1
    j += 1
  return res
```

```
str1 = "abcdefg"
str2 = "lllabcdlll"
res = algorithm(str1, str2, 0, 0)
print(res)
```
