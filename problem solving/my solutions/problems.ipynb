# Matrix - Minimum Path Sum

```
def min_path_sum_helper(matrix, m, n, path_visited_set, new_row, new_col, sum, path, direction, mem):
  cell_key = new_row * n + new_col
  if cell_key in path_visited_set:
    return None, None
  curr_path_visited_set = path_visited_set.copy()
  curr_path_visited_set.add(cell_key)
  curr_path = path.copy()
  curr_path.append(direction)
  return min_path_sum(matrix, m, n, curr_path_visited_set, new_row, new_col, sum + matrix[new_row][new_col], curr_path, mem)
```

```
def min_path_sum(matrix, m, n, path_visited_set, curr_row, curr_col, sum, path, mem):
  # memoization
  cell_key = curr_row * n + curr_col
  if cell_key in mem:
    return mem[cell_key]
  # base case
  if curr_row == m - 1 and curr_col == n - 1:
    return sum, path 
  # recursive step
  s1 = s2 = s3 = s4 = None
  # can go right
  if curr_col + 1 < n:
    s1, p1 = min_path_sum_helper(matrix, m, n, path_visited_set, curr_row, curr_col + 1, sum, path, "right", mem)
  # can go down
  if curr_row + 1 < m:
    s2, p2 = min_path_sum_helper(matrix, m, n, path_visited_set, curr_row + 1, curr_col, sum, path, "down", mem)
  # can go left
  if curr_col - 1 >= 0:
    s3, p3 = min_path_sum_helper(matrix, m, n, path_visited_set, curr_row, curr_col - 1, sum, path, "left", mem)
  # can go up
  if curr_row - 1 >= 0:
    s4, p4 = min_path_sum_helper(matrix, m, n, path_visited_set, curr_row - 1, curr_col, sum, path, "up", mem)
  # return min of paths

  if s1 == None and s2 == None and s3 == None and s4 == None:
    return None, None
  res = min(x for x in (s1, s2, s3, s4) if x)
  if res == s1:
    mem[cell_key] = [s1, p1]
    return s1, p1
  if res == s2:
    mem[cell_key] = [s2, p2]
    return s2, p2
  if res == s3:
    mem[cell_key] = [s3, p3]
    return s3, p3
  if res == s4:
    mem[cell_key] = [s4, p4]
    return s4, p4
```

```
def unique_paths(matrix, m, n, curr_row, curr_col, mem):
  # memoization
  cell_key = curr_row * n + curr_col
  if cell_key in mem:
    return mem[cell_key]
  # base case
  if curr_row == m - 1 and curr_col == n - 1:
    return 1
  # recursive step
  right_count = down_count = 0
  # can go right
  if curr_col + 1 < n:
    right_count = unique_paths(matrix, m, n, curr_row, curr_col + 1, mem)
  # can go down
  if curr_row + 1 < n:
    down_count = unique_paths(matrix, m, n, curr_row + 1, curr_col, mem)
  # return result
  mem[cell_key] = right_count + down_count
  return mem[cell_key]
```

```
def count_islands(matrix):
  # init algorithm
  m = len(matrix)
  n = len(matrix[0])
  island_map = {"island_key": "island_set"}
  # algorithm
  for row in range(0, m):
    for col in range(0, n):
      # found water
      if matrix[row][col] == 0:
        continue
      # found land
      cell_key = row * n + col
      up = (row - 1) * n + col
      left = row * n + (col - 1)
      # new island found
      if (not(up in island_map)) and (not(left in island_map)):
        print("new island")
        print(island_map[left])
        island_map[cell_key] = {cell_key}
        matrix[row][col] = cell_key
        continue
      # adjacent to island - up
      if (up in island_map) and (not(left in island_map)):
        print("up only")
        island_map[up].add(cell_key)
        matrix[row][col] = up
        continue
      # adjacent to island - left
      if (not (up in island_map)) and (left in island_map):
        print("left only")
        island_map[left].add(cell_key)
        matrix[row][col] = left
        continue
      # adjacent both
      if row >= m or row < 0 or col >= n or col < 0:
        continue
      # same island
      if matrix[row - 1][col] == matrix[row][col - 1]:
        print("same island - up and left")
        key = matrix[row - 1][col]
        island_map[key].add(cell_key)
        matrix[row][col] = key
        continue
      # up and left not same island - MERGE
      print("diff islands - up and left")
      left_island = island_map[left]
      # update curr cell
      island_map[up].add(cell_key)
      matrix[row][col] = up
      # update the other set
      del island_map[left]
      for cell in left_island:
        cell_col = cell % n
        cell_row = (cell - cell_col)//n
        island_map[up].add(cell)
        matrix[cell_row][cell_col] = up
  # return result
  return island_map
```

## **Matrix - Number of Islands (DFS)**

```
def dfs(matrix, island_key, island_set, visit_set, curr_row, curr_col):
  # init algorithm
  m = len(matrix)
  n = len(matrix[0])
  h = [0, 1, 0, -1]
  v = [1, 0, -1, 0]
  # can go left
  for k in range(0, 4):
    next_row = curr_row + h[k]
    next_col = curr_col + v[k]
    cell_key = next_row * n + next_col
    if next_row >= m or next_row < 0 or next_col >= n or next_col < 0:
      continue
    if (matrix[next_row][next_col] == 1) and not(cell_key in visit_set):
      visit_set.add(cell_key)
      island_set.add(cell_key)
      dfs(matrix, island_key, island_set, visit_set, next_row, next_col)
```

## **Matrix - Maximal Sqaure**

```
def create_acc_matrix(matrix):
  # init algorithm
  m = len(matrix)
  n = len(matrix[0])
  result = [ [0] * n for _ in range(m)]
  # get row acc
  for row in range(0, m):
    for col in range(0, n):
      # cell value = 0 
      if matrix[row][col] == 0:
        result[row][col] = [0, 0]
      # cell value = 1
      elif matrix[row][col] == 1:
        # edge cell
        if col == 0:
          result[row][col] = [1, 0]
        else:
          result[row][col] = [1 + result[row][col-1][0], 0]
  # get col acc
  for col in range(0, n):
    for row in range(0, m):
      # cell value = 0 
      if matrix[row][col] == 0:
        result[row][col][1] = 0
      # cell value = 1
      elif matrix[row][col] == 1:
        # edge cell
        if row == 0:
          result[row][col][1] = 1
        else:
          result[row][col][1] = 1 + result[row-1][col][1]
  # return result
  return result
```

```
import math
def algorithm(matrix):
  # init algorithm
  m = len(matrix)
  n = len(matrix[0])
  acc_matrix = create_acc_matrix(matrix)
  mem = [ [0] * n for _ in range(m)]
  maximal_square = {
      "area": 0,
      "row": -1,
      "col": -1
  }
  # traverse
  for row in range(0, m):
    for col in range(0, n):
      # found new possible square
      if matrix[row][col] == 1:
        # edge cell
        if row == 0 or col == 0:
          mem[row][col] = 1
        # non-edge cell
        else:
          left_acc = acc_matrix[row][col][0] - 1
          up_acc = acc_matrix[row][col][1] - 1
          prev_square = math.sqrt(mem[row-1][col-1])
          x = min(left_acc, up_acc, prev_square) + 1
          mem[row][col] = x * x
        # update max area
        my_area = mem[row][col]
        if my_area > maximal_square["area"]:
          maximal_square["area"] = my_area
          maximal_square["row"] = row
          maximal_square["col"] = col
  # return result
  return maximal_square
```

```
def create_acc_matrix(matrix):
  # init algorithm
  m = len(matrix)
  n = len(matrix[0])
  result = [ [0] * n for _ in range(m)]
  # get row acc
  for row in range(0, m):
    for col in range(0, n):
      # cell value = 0 
      if matrix[row][col] == 0:
        result[row][col] = [0, 0]
      # cell value = 1
      elif matrix[row][col] == 1:
        # edge cell
        if col == 0:
          result[row][col] = [1, 0]
        else:
          result[row][col] = [1 + result[row][col-1][0], 0]
  # get col acc
  for col in range(0, n):
    for row in range(0, m):
      # cell value = 0 
      if matrix[row][col] == 0:
        result[row][col][1] = 0
      # cell value = 1
      elif matrix[row][col] == 1:
        # edge cell
        if row == 0:
          result[row][col][1] = 1
        else:
          result[row][col][1] = 1 + result[row-1][col][1]
  # return result
  return result
```

```
import math
def algorithm(matrix):
  # init algorithm
  m = len(matrix)
  n = len(matrix[0])
  acc_matrix = create_acc_matrix(matrix)
  mem = [ [0] * n for _ in range(m)]
  maximal_rect = {
      "area": 0,
      "row": -1,
      "col": -1
  }
  # traverse
  for row in range(0, m):
    for col in range(0, n):
      # empty cell
      if matrix[row][col] == 0:
        mem[row][col] = [0, 0]
      # found new possible rectangle
      if matrix[row][col] == 1:
        # edge cell
        if row == 0 or col == 0:
          if col == 0:
            mem[row][col] = [1, acc_matrix[row][col][1]]
          else:
            mem[row][col] = [acc_matrix[row][col][0], 1]
        # non-edge cell
        else:
          left_acc = acc_matrix[row][col][0]
          up_acc = acc_matrix[row][col][1]
          prev_rectangle = mem[row-1][col-1]
          min_left = min(prev_rectangle[0]+1, left_acc)
          min_up = min(prev_rectangle[1]+1, up_acc)
          a1, a2, a3 = left_acc, up_acc, min_left * min_up
          if a1 >= a2 and a1 >= a3:
            mem[row][col] = [left_acc, 1]
          elif a2 >= a1 and a2 >= a3:
            mem[row][col] = [1, up_acc]
          elif a3 >= a1 and a3 >= a2:
            mem[row][col] = [min_left, min_up]
        # update max area
        my_area = mem[row][col][0] * mem[row][col][1]
        if my_area > maximal_rect["area"]:
          maximal_rect["area"] = my_area
          maximal_rect["row"] = row
          maximal_rect["col"] = col
  # return result
  return maximal_rect
```

## **BTS Permutations**

```
def algorithm(numbers):
  length = len(numbers)
  # base case
  if length == 1 or length == 0:
    return 1
  # recursive step
  length = len(numbers)
  res = 0
  for i in range(0, length):
    count_left = algorithm(numbers[0: i])
    count_right = algorithm(numbers[i+1: length])
    res += count_left * count_right
  # return result
  return res
```

```
def algorithm2(n):
  # init algorithm
  count = [0] * (n + 1)
  count[0] = count[1] = 1
  # bottom up
  for i in range(2, n + 1):
    for j in range(0, i):
      count[i] += count[i-j-1] * count[j]
  return count[n]
```

## **Binary Tree - Nodes at K Distance from given node**

```
def algorithm(node, target, k, result_set):
  # base case
  if node == target:
    search(node.left, 0, result_set, k)
    search(node.right, 0, result_set, k)
    return True, 0
  # recursive step
  # left subtree
  left_flag, left_len = algorithm(node.left, target, k, result_set)
  if left_flag == 1:
    if left_len + 1 == k:
      result_set.add(node)
      return 0, 0
    if left_len + 1 < k:
      search(node.right, left_len + 2, result_set, k)
      return True, left_len + 1
  # right subtree
  right_flag, right_len = algorithm(node.right, target, k, result_set)
  if right_flag == 1:
    if right_len + 1 == k:
      result_set.add(node)
      return 0, 0
    if right_len + 1 < k:
      search(node.left, right_len + 2, result_set, k)
      return True, right_len + 1
  # none of above i.e. node found in child but len exceeds k
  return 0, 0
```
