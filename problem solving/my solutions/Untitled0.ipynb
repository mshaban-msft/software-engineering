```
def algorithm(arr, k):
  n = len(arr)
  if n % k == 0:
    algorithm1(arr, k)
  else:
    algorithm2(arr, k)
```

```
def algorithm1(arr, k):
  n = len(arr)
  for start in range(0, k):
    prev = arr[start]
    curr = start + k
    while curr < n:
      tmp = arr[curr]
      arr[curr] = prev
      prev = tmp
      curr += k
    # initial positions
    curr = curr % n
    arr[curr] = prev
```

```
def algorithm2(arr, k):
  n = len(arr)
  counter = 1
  curr = 0
  prev = arr[curr]
  while counter <= k:
    next = curr + k
    if next > n:
      next = next % n
      counter += 1
    tmp = arr[next]
    arr[next] = prev
    prev = tmp
```
