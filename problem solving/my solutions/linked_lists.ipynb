## **Stack using an Array**

```
def Stack:
  # attributes
  elements = []
  max_size = 0
  stack_top = 0

  # constructor
  def __init__(self, size):
    self.elements = [None] * size
    self.max_size = size

  # interface methods
  def push(element):
    if self.stack_top + 1 < self.max_size:
      self.stack_top += 1 
      self.elements[stack_top] = element
    else:
      print('exceeded stack size!')

  def pop():
    if self.stack_top > 0:
      tmp = self.elements[stack_top]
      self.stack_top -= 1
      return tmp
    else:
      print('stack is empty!')

  def peek():
    if self.stack_top > 0:
      return self.elements[stack_top]
    else:
      print('stack is empty!')    

  def isEmpty():
    return self.stack_top >= 0
```

## **Queue using Array**

```
def Queue:
  # attributes
  elements = []
  head = 0
  tail = 0
  max_size = -1
  size = 0

  # constructor
  def __init__(self, size):
    self.elements = [None] * size
    self.max_size = size

  # methods
  def offer(value):
    if self.size == self.max_size:
      print("No Space!")
      return
    else:
      self.elements[self.tail] = value
      self.tail = (self.tail + 1) % self.max_size
      self.size += 1

  def pop():
    if self.size == self.max_size:
      print("No Space!")
      return
    else:
      tmp = self.elements[self.head]
      self.head = (self.head + 1) % self.max_size
      return tmp
```

## **Add two numbers**

```
def algorithm(l1, l2):
  n1 = l1.head
  n2 = l2.head
  result = LinkedList()
  carry = 0
  while (n1 is not None) and (l2 is not None):
    sum = n1.value + n2.value + carry
    carry = 0
    if sum % 10 != 0:
      carry = 1
    sum = sum // 10
    result.insert_last(sum)
    n1 = n1.next
    n2 = n2.next
  while n1 is not None:
    result.insert_last(n1.value + carry)
    carry = 0
  while n2 is not None:
    result.insert_last(n2.value + carry)
    carry = 0
  return result
```

## **LRU Cache**

```
class Node:
  # attributes
  value = None
  next = None
  prev = None

  # constructor
  def __init__(self, value):
    self.value = value
```

```
class DoublyHashLinkedList:
  # attributes
  head = None
  tail = None
  size = 0
  key_map = {}
  
  # constructor
  def __init__(self):
    pass

  # methods
  def insert_first(self, value):
    node = Node(value)
    if size == 0:
      self.head = node
      self.tail = node
    else:
      self.head.prev = node
      node.next = self.head
      self.head = node
    self.size += 1
    self.key_map[value] = node
  
  def insert_last(self, value):
    node = Node(value)
    if size == 0:
      self.head = node
      self.tail = node
    else:
      self.tail.next = node
      node.prev = self.tail
      self.tail = node
    self.size += 1
    self.key_map[value] = node

  def remove_node(self, value):
    if value in self.keys_map:
      node = self.keys_map[value]
      prev = node.prev
      next = node.next
      if prev == None and next == None: # only node in list
        self.head = self.tail = None
      if prev == None: # first node in list
        self.head = node.next
        self.head.prev = None
      if next == None: # last node in list
        self.tail = node.prev
        self.tail.next = None
      self.size -= 1
      del keys_map[value]
      return node

  def remove_first(self):
    if self.size == 0:
      return
    if self.size > 1:
      self.head.next.prev = None
      self.head = self.head.next
      self.size -= 1
    else:
      self.head = self.tail = None
      self.size -= 1

  def key_exists(self, value):
    return value in self.keys_map
```

```
class LRU:
  # attributes
  element_list = DoublyHashLinkedList()
  max_size = -1

  # constructor
  def __init__(self, max_size):
    self.max_size = max_size

  # methods
  def put(key, value):
    # key exists
    if key in self.element_list.key_exists(key):
      self.element_list.remove_node(key)
      self.element_list.insert_last(key)
    else:
      if self.size < self.max_size:
        self.element_list.insert_last(key)
      else:
        self.element_list.remove_first()
        self.element_list.insert_last(value)
```

### **Reverse Linked List - Recursive**

```
def reverse(node, prev):
  # base case
  if node.next == None:
    node.next = prev
    return node
  # recusive step
  new_head = reverse(node.next, node)
  node.next = prev
  return new_head
```
