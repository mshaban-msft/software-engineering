## **Set Matrix Zero**

```
def algorithm(matrix):
  # init
  m = len(matrix)
  n = len(matrix[0])
  cols = set(range(0, n))
  # traverse
  for row in range(0, m):
    for col in cols: # set iterator
      if matrix[row][col] == 0:
        set_zeroes(row, col, matrix, m, n) # set to zeroes
        cols.remove(col)  # remove col from set
        break # skip to next row
```
