## **Node Class**

```
class Node:
  # attributes
  value = None
  neighbors = []
  
  # contructor
  def __init__(self, value):
    self.value = value
```

## **BFS Search**

```
from collections import deque 

def BFS(start_node, value):
  # corner case checks
  # init algorithm
  queue = deque()
  queue.append(start_node)
  visit_set = {start_node}
  # algorithm body
  while len(queue) > 0:
    curr = queue.popleft()
    if curr.value == value:
      return True
    # expand children and add to visit set
    for neighbor in curr.neighbors:
      if neighbor in visit_set:
        continue
      queue.append(neighbor)
      visit_set.add(neighbor)
  # no node matched the search
  return False
```

## **DFS Search**

```
def DFS(start_node, value):
  # corner case checks
  # init algorithm
  stack = [start_node]
  visit_set = {start_node}
  # algorithm body
  while len(stack) > 0:
    curr = stack.pop()
    if curr.value == value:
      return True
    # expand children and add to visit set
    for neighbor in curr.neighbors:
      if neighbor in visit_set:
        continue
      stack.append(neighbor)
      visit_set.add(neighbor)
  # no node matched the search
  return False
```

### **Cycle in Undirected Graph**

```
def cycleDFS(node, prev, visit_set):
  # base case
  if node in visit_set:
    return True
  # recursive step
  visit_set.add(node)
  for neighbor in node.neighbors:
    if neighbor == prev:
      continue
    if cycleDFS(neighbor, node, visit_set):
      return True
  # nothing found
  return False
```
