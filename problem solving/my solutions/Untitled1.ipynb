```
def algorithm(str1, str2, p1, p2):
  # base case
  if p1 >= len(str1) and p2 < len(str2):
    return 0
  if p2 >= len(str2):
    return 1
  # recursive step
  if str1[p1] == str2[p2]:
    return algorithm(str1, str2, p1 + 1, p2) + algorithm(str1, str2, p1 + 1, p2 + 1)
  else:
    return algorithm(str1, str2, p1 + 1, p2)
```

```
str1 = "rabbbit"
str2 = "rabbit"
count = algorithm(str1, str2, 0, 0)
print(count)
```
